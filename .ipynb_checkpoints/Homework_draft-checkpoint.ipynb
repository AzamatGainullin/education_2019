```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def get_html(url):
    r = requests.get(url)
    return r.text
```

```python
def get_total_pages(html):
    soup = BeautifulSoup(html, 'lxml')
    pages = soup.find('div', class_='page__footer').find_all('a', class_='toggle-menu__item-link')
    total_pages = pages[-1].get('href').split('/')[-2][4:]
    return int(total_pages)
```

```python
def get_page_data(html):
    soup = BeautifulSoup(html, 'lxml')
    ads = soup.find('div', class_='posts_list').find_all('a', class_="post__title_link")
    spisok = []
    for ad in ads:
        
        try:
            title = ad.text
        except:
            title = ''
        try:
            url = ad.get('href')
        except:
            url = ''

        data = {'Название поста': title, 'URL-адрес': url}
        spisok.append(data)
        
        
    return spisok
```

```python
def main():
    url = 'https://habr.com/ru/users/olegchir/posts/page1/'
    base_url = 'https://habr.com/ru/users/olegchir/posts/page'
    total_pages = get_total_pages(get_html(url))
    spisok = []
    for i in range(1, total_pages+1):
        url_gen = base_url + str(i)
        html = get_html(url_gen)
        spisok.extend(get_page_data(html))
    return spisok
```

```python
posts_list_by_chiruhin = pd.DataFrame(main())
```

```python
for i in range(len(posts_list_by_chiruhin)):
    print(i)
    print(posts_list_by_chiruhin['Название поста'][i])
    print(posts_list_by_chiruhin['URL-адрес'][i])
```
